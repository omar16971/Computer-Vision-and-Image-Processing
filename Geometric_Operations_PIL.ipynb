```python
# donload the functions
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np
```

```python
#Helper function to plot images side by side
def plot_image(image_1, image_2,title_1="Orignal",title_2="New Image"):
    plt.figure(figsize=(10,10))
    plt.subplot(1, 2, 1)
    plt.imshow(image_1,cmap="gray")
    plt.title(title_1)
    plt.subplot(1, 2, 2)
    plt.imshow(image_2,cmap="gray")
    plt.title(title_2)
    plt.show()
```

## Geometric Transformtion

```python
# load the image
image = Image.open("lenna.png")
plt.imshow(image)
plt.show()
```

```python
# resize the width and height
width, height = image.size
new_width = 2 * width
new_hight = 2* height
new_image = image.resize((new_width, new_hight))
plt.imshow(new_image)
plt.show()
```

```python
# rotate the image
theta = 45
new_image = image.rotate(theta)
plt.imshow(new_image)
plt.show()
```

## Mathematical Operations

Before performing mathematical operations, the image has to be transformed to an array

```python
image = np.array(image)
```

```python
# adding contrast
new_image = image + 20
plt.imshow(new_image)
plt.show()
```

```python
# multipy the pixel intensity
new_image = image * 10
plt.imshow(new_image)
plt.show()
```

## Matrix operations

```python
image_gray = Image.open('baboon.png')
```

even thought the image is gray, it still has 3 channels. We need to convert it to one

```python
from PIL import ImageOps 
```

```python
image_gray = ImageOps.grayscale(image_gray)
plt.imshow(image_gray,cmap='gray')
plt.show()
```

```python
#use Singular Value Decomposition, decomposing our image matrix into a product of three matrices.
U, s, V = np.linalg.svd(image_gray, full_matrices = True)
```

```python
# we can see that s is not a rectangular matrix
s.shape
```

```python
S = np.zeros((image_gray.shape[0], image_gray.shape[1]))
S[:image.shape[0], :image.shape[0]] = np.diag(s)
```

```python
plot_image(U, V, title_1="Matrix U", title_2="Matrix V")
```

```python
plt.imshow(S, cmap='gray')
plt.show()
```

```python
B = S.dot(V)
plt.imshow(B,cmap='gray')
plt.show()
```
